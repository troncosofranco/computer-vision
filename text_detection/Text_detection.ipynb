```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import numpy as np
```

#2. Read image

```python
image_path = "/content/image2.png"
img = cv2.imread(image_path)
```

#3. Instance text detector

```python
#'en': english lenguage detection
reader = easyocr.Reader(['en'], gpu=False) 
```

#4. Detect text on image

```python
text_ = reader.readtext(img)
```

Define threshold

```python
threshold = 0.1
```

#5. Draw box and text

```python
for t_, t in enumerate(text_):
    print(t)

    bbox, text, score = t

    if score > threshold:
        cv2.rectangle(img, bbox[0], bbox[2], (0, 255, 0), 4)
        cv2.putText(img, text, bbox[0], cv2.FONT_HERSHEY_COMPLEX, 0.65, (255, 0, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
